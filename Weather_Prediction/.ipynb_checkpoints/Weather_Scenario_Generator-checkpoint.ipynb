```python
def in_bounding_box(point):
    """Determine whether a point is in our downtown bounding box"""
    lng, lat = point
    in_lng_bounds = DOWNTOWN_BOUNDING_BOX[0] <= lng <= DOWNTOWN_BOUNDING_BOX[2]
    in_lat_bounds = DOWNTOWN_BOUNDING_BOX[1] <= lat <= DOWNTOWN_BOUNDING_BOX[3]
    return in_lng_bounds and in_lat_bounds

def haversine(lonlat1, lonlat2):
    """
    Calculate the great circle distance in kilometers between two points 
    on the earth (specified in decimal degrees)
    """
    lon1, lat1 = lonlat1
    lon2, lat2 = lonlat2
    
    # convert decimal degrees to radians 
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])

    # haversine formula 
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a)) 
    r = 3440.064795 # Radius of earth in meters. Use 3956 for miles. Determines return value units.
    return c * r

```

```python
def return_model_predictor():
    clfs = []
    arrs = []
    for ahead in range(max_ahead):
        clf = load(f'LR_Models/model_{ahead}.joblib') 
        arr = np.loadtxt(f"LR_Models/cutoffs_{ahead}.csv", delimiter=",", dtype=float)
        
        clfs.append(clf)
        arrs.append(arr)
    
    def lr_prediction(node_data):
        Time_Nodes = (max_ahead)*Disc+1
        
        node_pred = np.zeros((Time_Nodes))
        
        k = 0
        for ahead in range(max_ahead):
            for d in range(Disc):
                prob = clfs[ahead].predict_proba(node_data.reshape(-1,1).T)[:,1][0]
                y_interp = interp1d(arr[:,2], arr[:,1])
                prob = y_interp(prob)
                node_pred[k] = prob
                k = k + 1
                
        return node_pred
    

    return lr_prediction
                
```

```python
lr_prediction = return_model_predictor()
```

```python
def save_scen_oracle(cutoff_ind, num, ind):
    
    pred_name = "Oracle"
    
    truth = Y_data[ind]
    wind = Winds[ind]
    
    Time_Nodes = (max_ahead)*Disc+1
     
    colors = []
    pred_vals = []
    
    node_truth = np.zeros((Time_Nodes, n_nodes))
    
    for n in range(n_nodes):
        
        x_array = []
        y_array = []
        
        for ahead in range(max_ahead):
            x_array.append((ahead)*Disc)
            y_array.append(truth[n,ahead])

            
        xv = np.array(x_array)
        yv = np.array(y_array)
        
        for i in range(Time_Nodes):
            pred_val = np.interp(i,xv,yv)
            node_truth[i,n] = pred_val
            
    Thres_Vals = np.zeros((20, 3))+0.5
    

    if os.path.exists(f"Weather_Scenarios/Weather_Cutoff_{cutoff_ind}"):
        None
    else:
        os.mkdir(f"Weather_Scenarios/Weather_Cutoff_{cutoff_ind}")

    
    file_name = f"Weather_Scenarios/Weather_Cutoff_{cutoff_ind}/Weather_Scenario_{num}"
    os.mkdir(file_name)
    os.mkdir(f"{file_name}/{pred_name}")
    
    
    np.savetxt(f"{file_name}/Weather_Scenario_Truth.csv", node_truth, delimiter=",")
    
    node_pred = 1-node_truth
    np.savetxt(f"{file_name}/{pred_name}/Weather_Scenario_Pred.csv", node_pred, delimiter=",")
    
    os.mkdir(f"{file_name}/{pred_name}/LonLats")
    for t in range(1):
        np.savetxt(f"{file_name}/{pred_name}/LonLats/Station_Lon_Lat_{t+1}.csv", LonLat_Array, delimiter=",")
```

```python
import pymap3d as pm
def save_scen_dynamic(cutoff_ind, num, ind):
    
    pred_name = "Dynamic"
    
    truth = Y_data[ind]
    wind = Winds[ind]
    
    Time_Nodes = (max_ahead)*Disc+1
     
    colors = []
    pred_vals = []
    
    node_pred = np.zeros((Time_Nodes, n_nodes))
    
    for n in range(n_nodes):
        
        x_array = []
        y_array = []
        
        for ahead in range(max_ahead):
            x_array.append((ahead)*Disc)
            y_array.append(truth[n,ahead])

            
        xv = np.array(x_array)
        yv = np.array(y_array)
        
        for i in range(Time_Nodes):
            pred_val = np.interp(0,xv,yv)
            node_pred[i,n] = pred_val
            
    Thres_Vals = np.zeros((20, 3))+0.5


    if os.path.exists(f"Weather_Scenarios/Weather_Cutoff_{cutoff_ind}"):
        None
    else:
        os.mkdir(f"Weather_Scenarios/Weather_Cutoff_{cutoff_ind}")
    
    file_name = f"Weather_Scenarios/Weather_Cutoff_{cutoff_ind}/Weather_Scenario_{num}"
    os.mkdir(f"{file_name}/{pred_name}")
    
    node_pred = 1-node_pred
    np.savetxt(f"{file_name}/{pred_name}/Weather_Scenario_Pred.csv", node_pred, delimiter=",")
    
    os.mkdir(f"{file_name}/{pred_name}/LonLats")
    for t in range(Time_Nodes):
        NewLonLat = deepcopy(LonLat_Array)
        for n in range(n_nodes):
            OrgLon = LonLat_Array[n,0]
            OrgLat = LonLat_Array[n,1]
            
            NewCord = pm.enu2geodetic(wind[n,0]*t,wind[n,1]*t,0, OrgLat, OrgLon, 0)
            
            NewLon = NewCord[1]
            NewLat = NewCord[0]
            
            NewLonLat[n,0] = NewLon
            NewLonLat[n,1] = NewLat
            
        np.savetxt(f"{file_name}/{pred_name}/LonLats/Station_Lon_Lat_{t+1}.csv", NewLonLat, delimiter=",")
```

```python
def save_scen_lr(cutoff_ind, num, ind):
    
    pred_name = "LinearRegression"
    
    truth = Y_data[ind]
    wind = Winds[ind]
    
    Time_Nodes = (max_ahead)*Disc+1
     
    colors = []
    pred_vals = []
    
    node_pred = np.zeros((Time_Nodes, n_nodes))
    
    for n in range(n_nodes):
        node_pred[:,n] = lr_prediction(X_data[ind][n,:])


    if os.path.exists(f"Weather_Scenarios/Weather_Cutoff_{cutoff_ind}"):
        None
    else:
        os.mkdir(f"Weather_Scenarios/Weather_Cutoff_{cutoff_ind}")
    
    file_name = f"Weather_Scenarios/Weather_Cutoff_{cutoff_ind}/Weather_Scenario_{num}"
    os.mkdir(f"{file_name}/{pred_name}")
    
    np.savetxt(f"{file_name}/{pred_name}/Weather_Scenario_Pred.csv", node_pred, delimiter=",")
    
    os.mkdir(f"{file_name}/{pred_name}/LonLats")
    np.savetxt(f"{file_name}/{pred_name}/LonLats/Station_Lon_Lat_1.csv", LonLat_Array, delimiter=",")
```
