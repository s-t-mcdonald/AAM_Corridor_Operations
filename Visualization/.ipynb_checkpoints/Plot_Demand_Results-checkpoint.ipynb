```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.set(font_scale = 2)
```

# Download Demand Data

```python
param_array = pd.read_csv("../parameter_array.csv")
param_array["param_num"] = param_array["PARAM"]


demand_data = pd.read_csv("../Results/demand_requests.csv")
demand_data["travel_time_reduction"] = demand_data["Car_TT"] - demand_data["Flying_TT"]
demand_data["param_num"] = param_array["PARAM"]
demand_data = demand_data.groupby(["param_num"]).agg(["mean","count","sum", "min", "max"])
demand_data = pd.merge(demand_data, param_array, how="left", on="param_num")

raw_demand_df = pd.read_csv("../Results/demand_requests.csv")
raw_demand_df["travel_time_reduction"] = raw_demand_df["Car_TT"] - raw_demand_df["Flying_TT"]
raw_demand_df["param_num"] = param_array["PARAM"]
raw_demand_df = pd.merge(raw_demand_df, param_array, how="left", on="param_num")
```

```python
demand_data.columns
```

# Plot Data

```python
TR_vals = np.linspace(0,100,100)
Alpha_Vals = [0.1, 0.075, 0.05]
for alpha in Alpha_Vals:
    Prob_Vals = 1/(1+np.exp(-alpha*(TR_vals-50)))
    plt.plot(TR_vals, Prob_Vals)
fig = plt.gcf()
fig.set_size_inches(10, 8)
plt.grid()
plt.legend([rf'$\alpha$ = {a}' for a in Alpha_Vals],fontsize=18)
plt.xlabel("Relative Time Reduction (%)", fontsize=18)
plt.ylabel(r'Probability of Demand (value*$\gamma$)', fontsize=18)
```

```python
demand_data["Demand_per_Min"] = demand_data[('demand_request', 'count')]/demand_data["HORIZON"]
gfg = sns.scatterplot(x="DEMAND_ALPHA",y="Demand_per_Min",hue="DEMAND_MAX", data=demand_data,s=200)
# plt.setp(gfg.get_legend().get_texts(), fontsize='18') 
# gfg.set_ylabel("Demand per Min", fontsize=20)
# gfg.set_xlabel(r"Demand Parameter - $\alpha$", fontsize=20)
```

```python
demand_data["Average Relative Travel Time Reduction"] = demand_data[('Time_Red', 'mean')]
gfg = sns.barplot(x="DEMAND_ALPHA",y="Average Relative Travel Time Reduction", data=demand_data)
gfg.set_ylabel("Relative Travel Time Reduction (%)")
```

```python
demand_data["Average Travel Time Reduction"] = demand_data[('travel_time_reduction', 'mean')]
gfg = sns.barplot(x="DEMAND_ALPHA",y="Average Travel Time Reduction", data=demand_data)
gfg.set_ylabel("Travel Time Reduction (min)")
```

```python
demand_data["Demand_Satisfied"] = demand_data[('RMP_CG', 'mean')]
gfg = sns.scatterplot(x="DEMAND_ALPHA",y="Demand_Satisfied",hue="DEMAND_MAX", data=demand_data,s=200)
gfg.set_ylabel("Percentatge of Demand Satisfied (%)")
```

```python
raw_demand_df["Demand Satisfied"] = np.abs(np.round(raw_demand_df["RMP_CG"]))
gfg = sns.barplot(x="DEMAND_ALPHA", y="Time_Red", hue="Demand Satisfied", data=raw_demand_df)
gfg.set_ylabel("Travel Time Reduction (min)")
```

```python
gfg = sns.barplot(x="DEMAND_ALPHA", y="Revenue", hue="Demand Satisfied", data=raw_demand_df)
gfg.set_ylabel("Revenue ($)")
```
