```python
import numpy as np
import networkx as nx
import json
import matplotlib.pyplot as plt
import random
import pyproj

import utm
import numpy as np
import random
import math
import time

import pandas as pd
import pydeck as pdk

import pydeck as pdk
import pandas as pd
import ipywidgets
# use palettable for colormap colors
from palettable.cartocolors.sequential import BrwnYl_3
import json

Colors_RGB = [[31, 119, 180],
              [227, 119, 194],
              [44, 160, 44],
              [255, 127, 14],
              [148, 103, 189],
              [140, 86, 75],
              [127, 127, 127],
              [188, 189, 34],
              [23, 190, 207]]
          
    
JFK_LAT_LON = 40.645944, -73.784839
```

```python
experiment_num = 1
result_num = 200
```

```python
g = nx.read_gml(f"../Practical_Results/Experiment_{experiment_num}/Results_{result_num}/Description/Graph.gml")

with open(f"../Practical_Results/Experiment_{experiment_num}/Results_{result_num}/Description/Truth_Closures.csv") as file_name:
    Truth_Closures = np.loadtxt(file_name, delimiter=",")

with open(f"../Practical_Results/Experiment_{experiment_num}/Results_{result_num}/Description/Visualization_Desc.json") as f:
        Visualization_Desc = json.load(f)
    
LongLat = nx.get_node_attributes(g, "pos")

geodesic = pyproj.Geod(ellps='WGS84')

n_nodes = g.number_of_nodes()

lons = []
lats = []
for n in range(n_nodes):
    lons.append(LongLat[str(n)][0])
    lats.append(LongLat[str(n)][1])
    
lons = np.asarray(lons)
lats = np.asarray(lats)

mult = 10

Num_Vertiports = Visualization_Desc["Num_Vertiports"]
min_time = 0
max_time = Visualization_Desc["Max_Time"]-1
max_time = max_time*mult
```

```python
SRC_list = []
DST_list = []

for e in g.edges:
    SRC_list.append([lons[int(e[0])], lats[int(e[0])]])
    DST_list.append([lons[int(e[1])], lats[int(e[1])]])
    
dictv = {'start': SRC_list, 'end': DST_list} 

edge_df = pd.DataFrame(dictv)
    
```

```python
WEATHER_PRED = ["Naive", "Predict", "Truth"]

Routes = {}
for weather_pred in WEATHER_PRED:


    with open(f"../Practical_Results/Experiment_{experiment_num}/Results_{result_num}/Description/Active_Routes_{weather_pred}.json") as f:
        Active_Routes = json.load(f)



    Index = []
    Paths = []
    Times = []
    Lanes = []
    Current_Location = []
    Colors = []
    Tooltip = []

    j = 0
    veh_df = {}

    

    for key in Active_Routes.keys():

        Desc = Active_Routes[key]
        l = Desc["Lanes"]
        origin = int(Desc["Path"][0]) - 1
        dst = int(Desc["Path"][-1]) - 1
        CG = Desc["CG"]
        transport = Desc["Transport"]

        path = [(lons[int(p) - 1], lats[int(p) - 1],l*5000) for p in Desc["Path"]]
        path.insert(0, (lons[origin]-0.001, lats[origin]-0.001,0))
        path.append((lons[dst]-0.001, lats[dst]-0.001,0))


        times = Desc["Times"]
        times = [t * mult for t in times]
        
        times.insert(0, times[0]-1)
        times.append(times[-1]+1)
        
        if transport == 0:
            color = [214, 39, 40]
        
        else:
            color = Colors_RGB[j % len(Colors_RGB)]
            j = j + 1

        Paths.append(path)
        Times.append(times)
        Colors.append(color)
        Lanes.append(l)
        if transport == 1:
            Time_Red = round(Desc["Time_Red"])
            Prob = Desc["Prob"]
            Tooltip.append(f"<b>Origin:</b> {origin} | <b>Destination:</b> {dst} | <b>Lane:</b> {l} | <b>CG:</b> {CG} | <b>Transport:</b> {transport} | Demand Travel Time Reduction {Time_Red}%")
        else:
            Tooltip.append(f"<b>Origin:</b> {origin} | <b>Destination:</b> {dst} | <b>Lane:</b> {l} | <b>CG:</b> {CG} | <b>Transport:</b> {transport}")

        


    dictv = {'path': Paths, 'time': Times, 'color': Colors, 'lane': Lanes, 'tooltip': Tooltip} 

    route_df = pd.DataFrame(dictv)


    Routes[weather_pred] = route_df
```

```python
full_df = {}

for k_val in range(-2,max_time):
    
    Node_Lon_Lat = []
    Node_Color = []
    Size = []
    Vertiport = []
    Tooltip = []
    
    k = max([k_val,0])

    for i in range(len(lons)):

        if (i < Num_Vertiports):
            Node_Lon_Lat.append([lons[i], lats[i]])
            Size.append(30)
            Tooltip.append(f"<b>Type:</b> Vertiport: {i}")
            
            if (Truth_Closures[int(k/mult),i] < 0.5):
                Node_Color.append([255, 0, 0, 255])
            elif (Truth_Closures[int(k/mult),i] > 0.5):
                Node_Color.append([255, 0, 0, 255])

            
        if (i >= Num_Vertiports):
            
            if len(list(nx.neighbors(g,str(i)))) > 2:
                Node_Lon_Lat.append([lons[i], lats[i]])
                Size.append(20)
                Tooltip.append(f"<b>Type:</b> Node: {i}")
                
                if (Truth_Closures[int(k/mult),i] < 0.5):
                    Node_Color.append([0, 255, 0, 255])
                elif (Truth_Closures[int(k/mult),i] > 0.5):
                    Node_Color.append([255, 0, 0, 255])

        
            


    dictv = {'coordinates': Node_Lon_Lat, 'color': Node_Color, 'size': Size, 'tooltip': Tooltip} 

    df = pd.DataFrame(dictv)
    
    full_df[k_val+1] = df
```

```python

sunlight = {
    "@@type": "_SunLight",
    "timestamp": 1564696800000,  # Date.UTC(2019, 7, 1, 22),
    "color": [255, 255, 255],
    "intensity": 1.0,
    "_shadow": True,
}

ambient_light = {"@@type": "AmbientLight", "color": [255, 255, 255], "intensity": 1.0}

lighting_effect = {
    "@@type": "LightingEffect",
    "shadowColor": [0, 0, 0, 0.5],
    "ambientLight": ambient_light,
    "directionalLights": [sunlight],
}

view_state = pdk.ViewState(
    **{"latitude": JFK_LAT_LON[0], "longitude": JFK_LAT_LON[1], "zoom": 6, "maxZoom": 14, "pitch": 30, "bearing": 0}
)


Trip_layer = pdk.Layer(
    type="TripsLayer",
    data=Routes["Naive"],
    get_path="path",
    get_timestamps="time",
    get_color="color",
    opacity=0.8,
    width_min_pixels=10,
    rounded=True,
    trail_length=7*mult,
    current_time=1,
    pickable=True,
)



layer = pdk.Layer(
    "ScatterplotLayer",
    full_df[1],
    pickable=True,
    stroked=True,
    filled=True,
    radius_scale=200,
    radius_min_pixels=1,
    radius_max_pixels=100,
    line_width_min_pixels=1,
    get_position="coordinates",
    get_radius="size",
    get_fill_color="color",
    get_line_color=[0, 0, 0],
)

line_layer = pdk.Layer(
    "LineLayer",
    edge_df,
    get_source_position="start",
    get_target_position="end",
    get_color=[105,105,105],
    get_width=0.6,
    highlight_color=[255, 255, 0],
    picking_radius=10,
    auto_highlight=True,
)

tooltip = {"html": "{tooltip}"}

keys = {"mapbox": "pk.eyJ1IjoibWNkb25zdCIsImEiOiJjbDRyNGphOWwweHA1M2lxemI1ZXhmM3Y1In0.Km2Tt6dKpTqwfNmTMEKrQA"}


layers = [Trip_layer]#[line_layer, layer, Trip_layer]

r = pdk.Deck(
    layers,
    initial_view_state=view_state,
    effects=[lighting_effect],
    map_style="dark",
    api_keys=keys,
    tooltip=tooltip,
)


time_slider = ipywidgets.IntSlider(value=-1, min=-1, max=max_time-1, step=1)

time_play = ipywidgets.Play(value=-1, min=-1, max=max_time-1, step=1, description='Press play', interval=int(200/mult))
ipywidgets.jslink((time_play, 'value'), (time_slider, 'value'))

Weather_Predict_Button = ipywidgets.ToggleButtons(
    options=["Naive", "Truth"],
    description='CG:',
    disabled=False,
    button_style=''
)


layout1 = ipywidgets.HBox([time_slider, time_play])
layout2 = ipywidgets.HBox([Weather_Predict_Button])

# # function

def update_graph(time, weather_pred):

    Trip_layer.data = Routes[weather_pred]
        
    
    Trip_layer.current_time = time
    
    layer.data = full_df[time]
    
    
    return r.update()


# # interaction between widget and function
interact = ipywidgets.interactive_output(update_graph, {'time': time_slider, 'weather_pred': Weather_Predict_Button});
display(layout1,layout2,interact)

# # display and save map (to_html(), show())
r.show()
```
