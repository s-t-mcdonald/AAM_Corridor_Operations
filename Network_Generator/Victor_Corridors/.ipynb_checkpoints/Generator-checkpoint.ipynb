```python

def is_in(lat,lon,UR,LL):
    return (lat >= LL[0]) and (lat <= UR[0]) and (lon >= LL[1]) and (lon <= UR[1])

def ccw(A,B,C):
    return (C[1]-A[1]) * (B[0]-A[0]) > (B[1]-A[1]) * (C[0]-A[0])

# Return true if line segments AB and CD intersect
def intersect(A,B,C,D):
    return ccw(A,C,D) != ccw(B,C,D) and ccw(A,B,C) != ccw(A,B,D)

def haversine(lon1, lat1, lon2, lat2):
    """
    Calculate the great circle distance in kilometers between two points 
    on the earth (specified in decimal degrees)
    """
    # convert decimal degrees to radians 
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])

    # haversine formula 
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a)) 
    r = 6371000 # Radius of earth in meters. Use 3956 for miles. Determines return value units.
    return c * r

def angle_of_vectors(a,b):


    inner = np.inner(a, b)
    norms = LA.norm(a) * LA.norm(b)

    cos = inner / norms
    rad = np.arccos(np.clip(cos, -1.0, 1.0))
    deg = np.rad2deg(rad)
    
    return deg

```

```python
Num_Vertiports = len(vertiport_locations)

lons = []
lats = []
Xs = []
Ys = []
New_Long_Lat = {}
i = 0

for i in range(Num_Vertiports):
    
    lon = vertiport_locations["geometry_object"][i].x
    lat = vertiport_locations["geometry_object"][i].y
    
    x, y, z, zl = utm.from_latlon(lat, lon)
    
    lons.append(lon)
    lats.append(lat)

    Xs.append(x)
    Ys.append(y)
    g.add_node(i)
    
    
    New_Long_Lat[i] = [lon, lat]
    
    i += 1
    
    
```

```python
nav_points = []
for nav in navs["features"]:
    lon, lat, alt = nav["geometry"]["coordinates"]
    
    if is_in(lat,lon,UR,LL):
        nav_points.append((lon,lat))
```

```python
i = Num_Vertiports
for nav in nav_points:
    lon = nav[0]
    lat = nav[1]
    
    Flag = False
    
    for j in range(Num_Vertiports,len(lons)):
        dist = ((lon-lons[j])**2 + (lat-lats[j])**2)**0.5
        if dist < 0.001:
            Flag = True
            
    if Flag:
        break
        
    
    x, y, z, zl = utm.from_latlon(lat, lon)
    
    lons.append(lon)
    lats.append(lat)

    Xs.append(x)
    Ys.append(y)
    g.add_node(i)
    
    
    New_Long_Lat[i] = [lon, lat]
    
    i += 1
```

```python
distmx = np.zeros((len(New_Long_Lat), len(New_Long_Lat)))
for i in range(Num_Vertiports,len(New_Long_Lat)):
    for j in range(Num_Vertiports,len(New_Long_Lat)):
        dist = ((lons[i]-lons[j])**2 + (lats[i]-lats[j])**2)**0.5
        
        if dist > 1.5:
            continue
        
        distmx[i,j] = dist
        g.add_edge(i,j)
```

```python
Flag = True

while Flag:
    
    Flag = False
    
    for e1 in g.edges:
        for e2 in g.edges:
            
            if not g.has_edge(e1[0],e1[1]):
                continue
                
            if not g.has_edge(e2[0],e2[1]):
                continue
            
            if (e1[0] in e2) or (e1[1] in e2):
                continue
            
                
            A = [lons[int(e1[0])], lats[int(e1[0])]]
            B = [lons[int(e1[1])], lats[int(e1[1])]]
            C = [lons[int(e2[0])], lats[int(e2[0])]]
            D = [lons[int(e2[1])], lats[int(e2[1])]]
                
            if intersect(A,B,C,D):
                if distmx[e1[0],e1[1]] > distmx[e2[0],e2[1]]:
                    g.remove_edge(e1[0],e1[1])
                else:
                    g.remove_edge(e2[0],e2[1])
                Flag = True
                break
            
            
                    
    
```

```python
Flag = True
while Flag:
    
    Flag = False
    
    for n in g.nodes:
#         print(n)
        nei = deepcopy(nx.neighbors(g, n))
        for n1 in nei:
            for n2 in nei:
                
                if not g.has_edge(n,n1):
                    continue
                
                if not g.has_edge(n,n2):
                    continue
                
                if n1 == n2:
                    continue
                
                A = np.array([lons[int(n)], lats[int(n)]])
                B = np.array([lons[int(n1)], lats[int(n1)]])
                C = np.array([lons[int(n2)], lats[int(n2)]])
                
                B = B - A
                C = C - A
                        
                ang = angle_of_vectors(B,C)
                
                if abs(ang) < 30:
    
                    
                    print(ang)
                    print(f'{n1} to {n2}')
                    
                    if distmx[n,n1] > distmx[n,n2]:
                        g.remove_edge(n,n1)
                    else:
                        g.remove_edge(n,n2)
                    Flag = True
                    
                    continue
                                    
                
```

```python
for i in range(Num_Vertiports):
    Min_Dist = 10**10
    Nav_Aid = 0
    for j in range(Num_Vertiports, len(New_Long_Lat)):
        dist = ((lons[i]-lons[j])**2 + (lats[i]-lats[j])**2)**0.5
        
        if dist < Min_Dist:
            Nav_Aid = j
            Min_Dist = dist
    
    g.add_edge(i,Nav_Aid)
```
