```python
import numpy as np
import geotable
import networkx as nx
import json
import spektral
from spektral.data import Graph, Dataset
import matplotlib.pyplot as plt
import tensorflow as tf
import random
import pyproj
import geojson
from haversine import haversine, Unit

import utm
import numpy as np
import random
import math
import time

import pandas as pd
import pydeck as pdk

import pydeck as pdk
import pandas as pd
import ipywidgets
# use palettable for colormap colors
from palettable.cartocolors.sequential import BrwnYl_3
import json

Colors_RGB = [[31, 119, 180],
              [255, 127, 14],
              [44, 160, 44],
              [214, 39, 40],
              [148, 103, 189],
              [140, 86, 75],
              [227, 119, 194],
              [127, 127, 127],
              [188, 189, 34],
              [23, 190, 207]]

    
JFK_LAT_LON = 40.645944, -73.784839



LL = 39.88494891-1, -75.33930212-1
UR = 41.888126+1, -72.73301048+3
```

```python
g = nx.Graph()
```

```python
vertiport_locations = geotable.load("../Vertiport_Locations/Vertiport_Locations.kmz")
with open("nav_aids.geojson", 'r') as f:
    navs = geojson.load(f)
```

```python

def is_in(lat,lon,UR,LL):
    return (lat >= LL[0]) and (lat <= UR[0]) and (lon >= LL[1]) and (lon <= UR[1])

def ccw(A,B,C):
    return (C[1]-A[1]) * (B[0]-A[0]) > (B[1]-A[1]) * (C[0]-A[0])

# Return true if line segments AB and CD intersect
def intersect(A,B,C,D):
    return ccw(A,C,D) != ccw(B,C,D) and ccw(A,B,C) != ccw(A,B,D)
```

```python
Num_Vertiports = len(vertiport_locations)

lons = []
lats = []
Xs = []
Ys = []
New_Long_Lat = {}
i = 0

for i in range(Num_Vertiports):
    
    lon = vertiport_locations["geometry_object"][i].x
    lat = vertiport_locations["geometry_object"][i].y
    
    x, y, z, zl = utm.from_latlon(lat, lon)
    
    lons.append(lon)
    lats.append(lat)

    Xs.append(x)
    Ys.append(y)
    g.add_node(i)
    
    
    New_Long_Lat[i] = [lon, lat]
    
    i += 1
    
    
```

```python
nav_points = []
for nav in navs["features"]:
    lon, lat, alt = nav["geometry"]["coordinates"]
    
    if is_in(lat,lon,UR,LL):
        nav_points.append((lon,lat))
```

```python
i = Num_Vertiports
for nav in nav_points:
    lon = nav[0]
    lat = nav[1]
    
    Flag = False
    
    for j in range(Num_Vertiports,len(lons)):
        dist = ((lon-lons[j])**2 + (lat-lats[j])**2)**0.5
        if dist < 0.001:
            Flag = True
            
    if Flag:
        break
        
    
    x, y, z, zl = utm.from_latlon(lat, lon)
    
    lons.append(lon)
    lats.append(lat)

    Xs.append(x)
    Ys.append(y)
    g.add_node(i)
    
    
    New_Long_Lat[i] = [lon, lat]
    
    i += 1
```

```python
distmx = np.zeros((len(New_Long_Lat), len(New_Long_Lat)))
for i in range(Num_Vertiports,len(New_Long_Lat)):
    for j in range(Num_Vertiports,len(New_Long_Lat)):
        dist = ((lons[i]-lons[j])**2 + (lats[i]-lats[j])**2)**0.5
        
        distmx[i,j] = dist
        g.add_edge(i,j)
```

```python
Flag = True

while Flag:
    
    Flag = False
    
    for e1 in g.edges:
        for e2 in g.edges:
            
            if not g.has_edge(e1[0],e1[1]):
                continue
                
            if not g.has_edge(e2[0],e2[1]):
                continue
            
            if (e1[0] in e2) or (e1[1] in e2):
                continue
            
                
            A = [lons[int(e1[0])], lats[int(e1[0])]]
            B = [lons[int(e1[1])], lats[int(e1[1])]]
            C = [lons[int(e2[0])], lats[int(e2[0])]]
            D = [lons[int(e2[1])], lats[int(e2[1])]]
                
            if intersect(A,B,C,D):
                if distmx[e1[0],e1[1]] > distmx[e2[0],e2[1]]:
                    g.remove_edge(e1[0],e1[1])
                else:
                    g.remove_edge(e2[0],e2[1])
                Flag = True
                break
                    
    
```

```python
for i in range(Num_Vertiports):
    Min_Dist = 10**10
    Nav_Aid = 0
    for j in range(Num_Vertiports, len(New_Long_Lat)):
        dist = ((lons[i]-lons[j])**2 + (lats[i]-lats[j])**2)**0.5
        
        if dist < Min_Dist:
            Nav_Aid = j
            Min_Dist = dist
    
    g.add_edge(i,Nav_Aid)
```

```python
SRC_list = []
DST_list = []

for e in g.edges:
    SRC_list.append([lons[int(e[0])], lats[int(e[0])]])
    DST_list.append([lons[int(e[1])], lats[int(e[1])]])
    
dictv = {'start': SRC_list, 'end': DST_list} 

edge_df = pd.DataFrame(dictv)
    
```
