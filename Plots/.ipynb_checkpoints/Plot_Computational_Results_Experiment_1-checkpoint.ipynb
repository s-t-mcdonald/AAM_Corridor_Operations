```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
import seaborn as sns
from os.path import exists
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.set(font_scale = 2)
%matplotlib inline
```

```python
import re
 
# helper function to perform sort
def num_sort(test_string):
    return list(map(int, re.findall(r'\d+', test_string)))[0]
```

# Download and Aggergate Data

```python
DIRS = os.listdir("../Computational_Results/Experiment_1/")
param_array = pd.read_csv("../Parameters/parameter_array_computational_1.csv")
param_array["param_num"] = param_array["PARAM"]
```

```python
data_frames = []
for d in DIRS:
    try:
        data_frames.append(pd.read_csv(f"../Computational_Results/Experiment_1/{d}/results_df.csv"))
        print(DataFrames)
    except:
        continue
agg_results_int_df = pd.concat(data_frames)


agg_results_int_df = pd.merge(agg_results_int_df, param_array, on="param_num")
# agg_results_int_df = agg_results_int_df[agg_results_int_df.HORIZON < 200]
agg_results_int_df = agg_results_int_df.reset_index()


agg_results_int_df["Gap"] = (agg_results_int_df.obj-agg_results_int_df.relaxed_objective)/agg_results_int_df.obj
```

```python
Max_Solve_Time = 1800
```

```python
# param_array = param_array.sort_values(by=['VERTIPORTS', 'NODE_MULT', 'DEMAND_PER_MIN'])

# param_array=param_array.astype(str)

LP_Solve = []
IP_Solve = []
IP_Solve_Times = []
IP_Solve_Val = []
Objective_Vals = []
IP_MIP_Gap = []
IP_TRUE_GAP = []
NAMES = []
for i in range(len(param_array)):
    
    Number_Vertiports = param_array.iloc[i].VERTIPORTS
    Node_Mult = int(param_array.iloc[i].NODE_MULT)
    Demand = int(param_array.iloc[i].DEMAND_PER_MIN)
    
    param_num = param_array.iloc[i].param_num
    
    Name = f"V-{Number_Vertiports}-{Node_Mult}-{Demand}"
    
    NAMES.append(Name)
    
    if exists(f"../Computational_Results/Experiment_1/Results_{param_num}/results_df.csv"):
        
        Solve_Time = agg_results_int_df[agg_results_int_df.param_num == param_num].solve_time.iloc[0]
        mip_gap = agg_results_int_df[agg_results_int_df.param_num == param_num].mip_gap.iloc[0]
        gap = agg_results_int_df[agg_results_int_df.param_num == param_num].Gap.iloc[0]
        obj = agg_results_int_df[agg_results_int_df.param_num == param_num].obj.iloc[0]
        
        if Solve_Time < 0:
            Solve_Time = Max_Solve_Time
            mip_gap = 1
            gap = 1
            obj = "N/A"

        IP_Solve_Times.append(Solve_Time)
        IP_MIP_Gap.append(mip_gap)
        IP_TRUE_GAP.append(gap)
        Objective_Vals.append(obj)
    
        if (agg_results_int_df[agg_results_int_df.param_num == param_num].relaxed_solve_time.iloc[0]) < 0:
            LP_Solve.append("No")
        else:
            LP_Solve.append("Yes")

        if (agg_results_int_df[agg_results_int_df.param_num == param_num].solve_time.iloc[0] > 0) & (agg_results_int_df[agg_results_int_df.param_num == param_num].mip_gap.iloc[0] < 0.02):
            
            IP_Solve.append("Yes")
            IP_Solve_Val.append(1)
            
            
        else:
            IP_Solve.append("No")
            IP_Solve_Val.append(0)
            
    else:

        LP_Solve.append("No")
        IP_Solve.append("No")
        IP_Solve_Times.append(Max_Solve_Time)
        IP_Solve_Val.append(0)
        IP_MIP_Gap.append(1)
        IP_TRUE_GAP.append(1)

        
param_array["IP_Solve"] = IP_Solve
param_array["LP_Solve"] = LP_Solve
param_array["Names"] = NAMES
param_array["IP_Solve_Times"] = IP_Solve_Times
param_array["IP_Solve_Val"] = np.array(IP_Solve_Val)*100
param_array["IP_MIP_Gap"] = np.array(IP_MIP_Gap)*100
param_array["IP_TRUE_GAP"] = np.array(IP_TRUE_GAP)*100
```

```python
param_array_names = param_array.groupby(["VERTIPORTS","NODE_MULT","DEMAND_PER_MIN"]).first()["Names"]
```

```python
param_array_names.to_csv("names.csv")
```
